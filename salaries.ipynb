```python
import matplotlib.pyplot as plt
import numpy as np

def calculate_monthly_take_home_salary(gross_annual):
    annual_take_home = 0
    if gross_annual <= 12570:
        annual_take_home = gross_annual
    elif gross_annual <= 50270:
        annual_take_home = 12570 + (gross_annual - 12570) * 0.8
    elif gross_annual <= 150000:
        annual_take_home = 12570 + (50270 - 12570) * 0.8 + (gross_annual - 50270) * 0.6
    else:
        annual_take_home = 12570 + (50270 - 12570) * 0.8 + (150000 - 50270) * 0.6 + (gross_annual - 150000) * 0.55
    return annual_take_home / 12

gross_salaries_annual = np.arange(10000, 160000, 10000)
monthly_take_home_salaries = np.array([calculate_monthly_take_home_salary(gross) for gross in gross_salaries_annual])

plt.plot(gross_salaries_annual, monthly_take_home_salaries, label='Monthly Take-Home vs Gross Annual Salary')
plt.xlabel('Gross Annual Salary (£)')
plt.ylabel('Monthly Take-Home Salary (£)')
plt.title('UK Monthly Take-Home Salaries vs Gross Annual Salaries')
plt.legend()
plt.grid()
plt.show()
```
